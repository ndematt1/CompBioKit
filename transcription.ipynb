```python
def transcribe(dna_seq, coding_strand = True): #define the function
    rna_seq = "" #make a string for rna seq
    
    #Error Handling. If the user does not put a boolean for coding strand it will end the function.
    if not isinstance(coding_strand, bool):
        return "Please enter True or False for the coding strand"
    
    #Error Handling. If the user tries to put something that you cannot capitilize
    #Will tell the user to enter a valid DNA sequence
    try:  
        dna_seq = dna_seq.upper()
    except AttributeError:
        return("Enter a valid DNA sequence") 
    
    #If the sequence is the coding strand. It will replace only T with U in order to transcribe to RNA
    if coding_strand: 
        for nuc in dna_seq:
            if nuc == "T":
                rna_seq += "U"
            elif nuc == "A":
                rna_seq += "A"
            elif nuc == "G":
                rna_seq += "G"
            elif nuc == "C":
                rna_seq += "C"
            else:
                return("Please enter a valid DNA sequence") #If the user puts any letter besides those found in nucleotides
                                                            #it will tell the user to put a valid DNA sequence     
        return rna_seq
            

    elif coding_strand == False:  #If the user is  trying to transcibe the template strand
        for nuc in dna_seq:       #it will replace the nucleotide sequences with their complementary bases
            if nuc == "G":
                rna_seq += "C"
            elif nuc == "C":
                rna_seq +="G"
            elif nuc == "A":
                rna_seq +="U"
            elif nuc == "T":
                rna_seq +="A"
            else: 
                return("Enter a valid DNA sequence")
        return rna_seq
```

```python
def complement(dna_seq): #define function
    dna_comp_seq = ""
    
    
    #Error Handling. If the user tries to put something that you cannot capitilize
    #Will tell the user to enter a valid DNA sequence
    try: 
        dna_seq = dna_seq.upper()
    except AttributeError:
        return("Enter a valid DNA sequence")
    

#This code will make the DNA sequences complimentary DNA sequence
    for nuc in dna_seq:
        if nuc == "G":
            dna_comp_seq += "C"
        elif nuc == "C":
            dna_comp_seq +="G"
        elif nuc == "A":
            dna_comp_seq +="T"
        elif nuc == "T":
            dna_comp_seq +="A"
        else: 
            return("Enter a valid DNA sequence")
    return dna_comp_seq
```

```python
def reverse_complement(dna_seq): #define function
    
    #Error Handling. If the user tries to put something that you cannot capitilize
    #Will tell the user to enter a valid DNA sequence
    try: 
        dna_seq = dna_seq.upper()
    except AttributeError:
        return("Enter a valid DNA sequence")
    
    #Gets the complementary DNA sequence
    dna_comp_seq = complement(dna_seq)
   
    #Converts the sequence into a list
    comp_seq_ls = []
    for nuc in dna_comp_seq:
        comp_seq_ls.append(nuc)
    
    #Reverse the list
    rev_comp_seq_ls = comp_seq_ls
    rev_comp_seq_ls.reverse()

    #Turn the list into a string
    rev_comp_seq = "".join(rev_comp_seq_ls)
    return rev_comp_seq
    
```
