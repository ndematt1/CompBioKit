```python
def check_rna(rna_seq):
    
    rna_nuc_ls = ["U","A","G","C"]
    
    try:
        rna_seq = rna_seq.upper()
    except (AttributeError, NameError):
        return False
    

    for nuc in rna_seq:
        if nuc not in rna_nuc_ls:
            return False
    
    return True
    
```

```python
def check_dna(dna_seq):
    
    dna_nuc_ls = ["T","A","G","C"]
    
    try: 
        dna_seq = dna_seq.upper()
    except (AttributeError, NameError):
        return False
    

    for nuc in dna_seq:
        if nuc not in dna_nuc_ls:
            return False
    
    return True
```
