```python
def count_nuc(sequence):
   
    if check_dna(sequence) == True or check_rna(sequence) == True:
        nuc_count_dict = {"A":0, "T":0, "U":0, "G":0, "C":0}
        for nuc in sequence:
            nuc_count_dict[nuc] += 1
        return nuc_count_dict
    else:
        return("Please enter a valid nucleotide sequence")
```

```python
def gc_content(dna_seq):
    
    if check_dna(dna_seq) == False:
        return("Enter a valid DNA sequence")

    gc = dna_seq.count("G") + dna_seq.count("C")
    return gc / len(dna_seq)
```
