```python
seq = "AUGAUGAAAUUUGGGCCCUUAUAGAAAAAAAAAAUGUUUGGGUAAUUUUUUGGGCCCAAAAUGAUGUAA"

def translate(rna_seq, one_letter = True): #If one-letter is False it provides a three lettter abbreviation
    protein_ls = [] #create a list of amino acids in the protein 
    start_codon = False # Set the start codon to be false

    if not isinstance(one_letter, bool): # Error handling. If the argument does not have True or False.
        return "Please enter True or False for the one-letter abbreviation parameter"
    
    try: #Error Handling. If you cannot capitalize the sequence.
        rna_seq = rna_seq.upper()
    except AttributeError:
        return("Enter a valid RNA sequence")
    
    #List of valid nucleotides for RNA
    nuc_ls = ["U","A","G","C"]
    
    #If the nucleotide is not in the list. Enter a valid RNA sequence.
    for nuc in rna_seq:
        if nuc not in nuc_ls:
            return("Please enter a valid RNA sequence ")
    
    #For a nucleotide starting from index of 0 and ending at the last nucleotide in groups of 3
    for nuc in range(0, len(rna_seq), 3):
        codon = rna_seq[nuc:nuc + 3] #0-3

        #Break if the codon length is not equal to 3 
        if len(codon) != 3:
            break
        
        #If it is not the start codon. Dont start appending to the list.
        if not start_codon:
            if codon == "AUG":
                start_codon = True
        
        #If it is the stop codon break 
        if codon in ["UAA", "UAG", "UGA"]:
            break
        
        #while the start codon == True print either the one letter or three letter abbreviation
        if start_codon ==True:
            if one_letter:
                amino_acid = codon_dict[codon]
                protein_ls.append(amino_acid[0])
            
            elif one_letter == False:
                amino_acid = codon_dict[codon]
                protein_ls.append(amino_acid[1])

    #Make the list a string    
    if one_letter:
        one_letter_ls = "".join(protein_ls)
        return one_letter_ls
    
    elif one_letter == False:
        three_letter_ls = "-".join(protein_ls)
        return three_letter_ls
    

print(translate(seq, False))
```

```python
def translate_orfs(rna_seq, one_letter = False):
    ############################################################
    #Error Handling

    start_codon_index_ls = []
    stop_codon_index_ls = []
    total_protein_ls = []
    start_stop =[]


    if not isinstance(one_letter, bool): # Error handling. If the argument does not have True or False.
        return "Please enter True or False for the one-letter abbreviation parameter"

    try: #Error Handling. If you cannot capitalize the sequence.
        rna_seq = rna_seq.upper()
    except AttributeError:
        return("Enter a valid RNA sequence")

    #List of valid nucleotides for RNA
    nuc_ls = ["U","A","G","C"]

    #If the nucleotide is not in the list. Enter a valid RNA sequence.
    for nuc in rna_seq:
        if nuc not in nuc_ls:
            return("Please enter a valid RNA sequence ")
    #################################################################
    
    #Find all the start codons and put the index in a list
    for nuc in range(0, len(rna_seq), 3):
        codon = rna_seq[nuc:nuc + 3] #0-3
        if codon == "AUG":
            start_codon_index_ls.append(nuc)
    
    #Get all the stop codons and put the index in a list 
    for nuc in range(0, len(rna_seq), 3):
        codon = rna_seq[nuc:nuc + 3] #0-3
        if codon in ["UAA", "UAG", "UGA"]:
            stop_codon_index_ls.append(nuc)
    
    #last stop = -1, so the first start is always true
    last_stop = -1
    #for start codon in the start codon list
    for start in start_codon_index_ls:
    #if the start is less than the last stop continue
        if start < last_stop:
            continue
        
        #for stop in the stop codon list     
        for stop in stop_codon_index_ls:
            #if the stop is greater than the start
            if stop > start:
                    #translate
                    protein = translate(rna_seq[start:stop+3], one_letter)
                    total_protein_ls.append(protein)
                    #update the last stop postion
                    start_stop.append([start,stop])
                    last_stop = stop
                    break
            
    for i in range(len(start_stop)):
        print(f"ORF{i+1}: {start_stop[i]} -> {total_protein_ls[i]}")
            
    return(total_protein_ls)

```
